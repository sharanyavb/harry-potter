```python
hp1 = open("hp1.txt").read()
```

```python
hp1 = hp1.replace("\\", "")
hp1 = hp1.split("\n")
```

```python
empty = ""
page = "Page "
for text in hp1:
    if text == empty:
        hp1.remove(text)
for text in hp1:
    if page in text:
        hp1.remove(text)
```
