```python
# Import list of characters + info
characters = pd.read_csv("HP_Characters.csv")
count = characters["Full Intro"].count()

# Loop through dataframe, saving character info, scraping data, pushing to DB
for x in range(105, count):
    character = characters.iloc[x, 0]
    url = 'http://harrypotter.wikia.com/wiki/' + str(character)
    print(url)
    html = requests.get(url)
    soup = BeautifulSoup(html.text, 'html.parser')
    body = soup.body
    
    full_name = body.find("h2", class_="pi-item").text
    print(full_name)
    try:
        blood_status = body.find("h3", string="Blood status").next_sibling.next_sibling.text.split('[')[0]
        print(blood_status)
    except:
        blood_status = "N/A"
        print(blood_status)
    try:
        house = body.find("h3", string="House").next_sibling.next_sibling.text.split('[')[0]
        print(house)
    except:
        house = "N/A"
        print(house)
    try:
        species = body.find("h3", string="Species").next_sibling.next_sibling.text.split('[')[0]
        print(species)
    except:
        species = "Human"
        print(species)
    image = body.find("img", class_="pi-image-thumbnail")["src"]
    print(image)
    
    # Get other variables from CSV
    first_appearance = characters.iloc[x, 1]
    last_appearance = characters.iloc[x, 2]
    stone_first_appearance = characters.iloc[x, 5]
    stone_last_appearance = characters.iloc[x, 6]
    stone_year = characters.iloc[x, 7]
    chamber_first_appearance = characters.iloc[x, 8]
    chamber_last_appearance = characters.iloc[x, 9]
    chamber_year = characters.iloc[x, 10]
    prisoner_first_appearance = characters.iloc[x, 11]
    prisoner_last_appearance = characters.iloc[x, 12]
    prisoner_year = characters.iloc[x, 13]
    goblet_first_appearance = characters.iloc[x, 14]
    goblet_last_appearance = characters.iloc[x, 15]
    goblet_year = characters.iloc[x, 16]
    phoenix_first_appearance = characters.iloc[x, 17]
    phoenix_last_appearance = characters.iloc[x, 18]
    phoenix_year = characters.iloc[x, 19]
    prince_first_appearance = characters.iloc[x, 20]
    prince_last_appearance = characters.iloc[x, 21]
    prince_year = characters.iloc[x, 22]
    hallows_first_appearance = characters.iloc[x, 23]
    hallows_last_appearance = characters.iloc[x, 24]
    hallows_year = characters.iloc[x, 25]
    
    
    # Insert dictionary to databases
    db.full.insert_one({
                "character": character,
                "full_name": full_name,
                "blood_status": blood_status,
                "house": house,
                "species": species,
                "image_url": image,
                # "boggart": boggart,
                # "patronus": patronus,
                # "wand": wand,
                "first_appearance": first_appearance,
                "last_appearance": last_appearance
            })
    db.stone.insert_one({
                "character": character,
                "full_name": full_name,
                "blood_status": blood_status,
                "house": house,
                "species": species,
                "image_url": image,
#                 "boggart": boggart,
#                 "patronus": patronus,
#                 "wand": wand,
                "first_appearance": stone_first_appearance,
                "last_appearance": stone_last_appearance,
                "year": stone_year
            })
    db.chamber.insert_one({
                "character": character,
                "full_name": full_name,
                "blood_status": blood_status,
                "house": house,
                "species": species,
                "image_url": image,
#                 "boggart": boggart,
#                 "patronus": patronus,
#                 "wand": wand,
                "first_appearance": chamber_first_appearance,
                "last_appearance": chamber_last_appearance,
                "year": chamber_year
            })
    db.prisoner.insert_one({
                "character": character,
                "full_name": full_name,
                "blood_status": blood_status,
                "house": house,
                "species": species,
                "image_url": image,
#                 "boggart": boggart,
#                 "patronus": patronus,
#                 "wand": wand,
                "first_appearance": prisoner_first_appearance,
                "last_appearance": prisoner_last_appearance,
                "year": prisoner_year
            })
    db.goblet.insert_one({
                "character": character,
                "full_name": full_name,
                "blood_status": blood_status,
                "house": house,
                "species": species,
                "image_url": image,
#                 "boggart": boggart,
#                 "patronus": patronus,
#                 "wand": wand,
                "first_appearance": goblet_first_appearance,
                "last_appearance": goblet_last_appearance,
                "year": goblet_year
            })
    db.phoenix.insert_one({
                "character": character,
                "full_name": full_name,
                "blood_status": blood_status,
                "house": house,
                "species": species,
                "image_url": image,
#                 "boggart": boggart,
#                 "patronus": patronus,
#                 "wand": wand,
                "first_appearance": phoenix_first_appearance,
                "last_appearance": phoenix_last_appearance,
                "year": phoenix_year
            })
    db.prince.insert_one({
                "character": character,
                "full_name": full_name,
                "blood_status": blood_status,
                "house": house,
                "species": species,
                "image_url": image,
#                 "boggart": boggart,
#                 "patronus": patronus,
#                 "wand": wand,
                "first_appearance": prince_first_appearance,
                "last_appearance": prince_last_appearance,
                "year": prince_year
            })
    db.hallows.insert_one({
                "character": character,
                "full_name": full_name,
                "blood_status": blood_status,
                "house": house,
                "species": species,
                "image_url": image,
#                 "boggart": boggart,
#                 "patronus": patronus,
#                 "wand": wand,
                "first_appearance": hallows_first_appearance,
                "last_appearance": hallows_last_appearance,
                "year": hallows_year
            })
```
